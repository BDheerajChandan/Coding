```python
'''
z test - passport
1S -bollywood
2s - healthdrink
paired - breakups
Anova - onestop
chisquare - normal list
'''
#Z-Test
import pandas as pd
import math
data=pd.read_csv("passport.csv")
def z_test(pop_mean,pop_sd,sample):
    z_score=(sample.mean()-pop_mean)/(pop_sd/(math.sqrt(len(sample))))
    return z_score,stats.norm.cdf(z_score)
z_test(30,12.5,data["processing_time"])
```
