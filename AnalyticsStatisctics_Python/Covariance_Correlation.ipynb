# Covarianceand and Correlation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Random dataframe 
data=pd.DataFrame(np.random.randn(100,2),columns=["c1","c2"])
print(data)
# Finding covariance

cov_matrix=np.cov(data["c1"],data["c2"])
print("Covariance :\n",cov_matrix)

# Finding correlation
cor_matrix=np.corrcoef(data["c1"],data["c2"])
print("Correlation :\n",cor_matrix)

#plotting data
plt.bar(range(len(data["c1"])),data["c1"],label="c1")
plt.bar(range(len(data["c2"])),data["c2"],label="c2")
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot of c1 and c2')
plt.legend()
plt.grid(axis='both')
plt.plot()
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
sns.distplot(data["c1"],label="c1")
sns.distplot(data["c2"],label="c2")

plt.xlabel('Value')  # Label for x-axis
plt.ylabel('Density')  # Label for y-axis
plt.title('Distribution Plot of c1 and c2')
plt.legend()
plt.grid(axis='both')
```
