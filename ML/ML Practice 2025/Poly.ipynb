```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x
```

```python
y
```

# 4. Model Building

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=3)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(x_train_poly,y_train)
```

```python
import joblib
joblib.dump(model_lr,"Poly_model_lr.pickle")
load_poly_model_lr=joblib.load("Poly_model_lr.pickle")
```

# 5. Model Prediction

```python
y_pred_poly=load_poly_model_lr.predict(x_test_poly)
y_pred_poly
```

# 6. Model Evaluation

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
import numpy as np
print(mean_squared_error(y_test,y_pred_poly))
print(mean_absolute_error(y_test,y_pred_poly))
print(np.sqrt(mean_squared_error(y_test,y_pred_poly)))

print(load_poly_model_lr.score(x_test_poly,y_test))
print(load_poly_model_lr.score(x_train_poly,y_train))
#print(load_poly_model_lr.score(x_test,y_test))
#print(load_poly_model_lr.score(x_train,y_train))
```
