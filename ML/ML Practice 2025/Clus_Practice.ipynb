```python
import pandas as pd
data=pd.read_csv("Dataset_Files\\Employee_attrition.csv")
data
```

```python
fil_col=list(data.columns[data.dtypes=="object"])
fil_col
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
def f(i):
    data[i]=le.fit_transform(data[i])

for i in fil_col:
    f(i)

data
```

```python
data.info()
```

```python
from sklearn.cluster import KMeans
cl=3
km=KMeans(n_clusters=cl)
km.fit(data)
```

```python
cc=km.cluster_centers_
lbl=km.labels_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
cn=11
plt.scatter(data.iloc[:,0],data.iloc[:,cn],c=lbl,edgecolors="black")
plt.scatter(cc[:,0],cc[:,cn],label=f"{data.columns[cn]}_{cn+1}")
plt.show()
```

```python
sns.distplot(cc)
sns.distplot(lbl)
```
