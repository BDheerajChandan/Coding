# 3. Feature exttraction

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
print(x)
```

```python
print(y)
```

# 4. Model building

```python
# Split data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print("x_train shape : ",x_train.shape)
print("y_train shape : ",y_train.shape)
print("x_test shape : ",x_test.shape)
print("y_test shape : ",y_test.shape)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
# Train data
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

# 5. Model evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np
test_acc=model.score(x_test,y_test)
train_acc=model.score(x_train,y_train)

y_pred=model.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

print("Testing score : ",test_acc)
print("Training score : ",train_acc)
print("mae : ",mae)
print("mse : ",mse)
print("rmse : ",rmse)
```

# 6. Result visualization

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

x_set=sc.inverse_transform(x_train)
y_set=y_train

x1,x2=np.meshgrid(np.arange(x_set[:,0].min()-10,x_set[:,0].max()+10,0.25),
                  np.arange(x_set[:,1].min()-1000,x_set[:,1].max()+1000,0.25))

plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())

plt.contourf(x1,x2,model.predict(sc.transform(np.array([x1.ravel(),x2.ravel()]).T)).reshape(x1.shape),
             alpha=0.75,cmap=ListedColormap(('red','green')))

for i,j in enumerate(np.unique(x_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],c=ListedColormap(('red','green'))(i),label=j)
```
