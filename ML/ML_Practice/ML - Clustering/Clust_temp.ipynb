```python
import pandas as pd
data=pd.read_csv("C:\\Users\\KIIT\\Desktop\\ML Practice\\Dataset\\Machine Learning A-Z (Codes and Datasets)\\Part 4 - Clustering\\Section 24 - K-Means Clustering\\Python\\Mall_Customers.csv")
data.head()
```

```python
x=data.iloc[:,[3,4]].values
print(x)
```

```python
from sklearn.cluster import KMeans
wcss=list()
for i in range(1,11):
    model=KMeans(n_clusters=i,init='k-means++',random_state=42)
    model.fit_transform(x)
    wcss.append(model.inertia_)
```

```python
model_test=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans=model_test.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.scatter(range(1,11),wcss)
```

```python
ListColor=['r','g','b','c','m']

for i in range(5):
    plt.scatter(x[y_kmeans==i,0],x[y_kmeans==i,1],s=10,c=ListColor[i],label=f'Cluster {i}')
plt.scatter(model_test.cluster_centers_[:,0],model_test.cluster_centers_[:,1],s=100,c='y',label='centroid')
plt.legend()
plt.show()
```
