```python
import pandas as pd
data=pd.read_csv("C:\\Users\\KIIT\\Desktop\\ML Practice\\Dataset\\Machine Learning A-Z (Codes and Datasets)\\Part 4 - Clustering\\Section 25 - Hierarchical Clustering\\Python\\Mall_Customers.csv")
data.head()
```

```python
x=data.iloc[:,[3,4]].values
print(x)
```

```python
import scipy.cluster.hierarchy as sch
model=sch.dendrogram(sch.linkage(x,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
cl=['r','b','g','c','m']
for i in range(5):
    plt.scatter(x[y_hc==i,0],x[y_hc==i,1],c=cl[i],s=10,label=f'cluster_{i}')
plt.legend()
plt.show()
```
