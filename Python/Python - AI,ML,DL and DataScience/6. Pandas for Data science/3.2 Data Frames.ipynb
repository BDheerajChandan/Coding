```python
# Conditional selection

import numpy as np
import pandas as pd
from numpy.random import rand
np.random.seed(101)
df1=pd.DataFrame(rand(5,4),['A','B','C','D','E'],['C1','C2','C3','C4'])
print(df1)

print(df1>0.5,'\n')
bool_df1=df1>0.5
print(bool_df1,'\n')
print(df1[bool_df1],'\n')
print(df1['C1'],'\n')
print(df1['C1']>0.5,'\n')
print(df1[df1['C1']>0.5],'\n')
print(df1[df1['C1']>0.5]['C2'],'\n')

print(df1,'\n')
res1=df1[df1['C1']>0.5]
print(res1,'\n')
print(res1['C2'],'\n')
print(df1[df1['C1']<0.5],'\n')
print(df1[df1['C1']>0.5]['C2'],'\n')
print(df1[df1['C1']>0.5][['C2','C4']],'\n')

bool_df2=df1['C2']>0.5
res2=df1[bool_df2]
print(bool_df2,'\n')
print(res2,'\n')

print(res2[['C1','C3','C4']],'\n')
cols=['C1','C3','C4']
print(res2[cols],'\n')

print(df1,'\n')
print(df1[(df1['C1']>0.3) & (df1['C1']<0.7)],'\n')
print(df1[(df1['C1']>0.3) | (df1['C1']<0.7)],'\n')
print(df1[(df1['C1']>0.3) & (df1['C3']<0.7)],'\n')
print(df1[(df1['C1']>0.3) | (df1['C3']<0.7)],'\n')
```

```python
# .reset_index() - The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.
#                  Pandas reset_index() is a method to reset index of a Data Frame. reset_index() method sets a list of integer 
#                  ranging from 0 to length of data as index.
# Pandas set_index() is a method to set a List, Series or Data frame as index of a Data Frame.
#                   set_index() is used to set the index to pandas DataFrame.
print(df1,'\n')
print(df1.reset_index(),'\n')

newind="1 2 3 4 5".split()
print(newind)
df1['C5']=newind
print(df1,'\n')


print(df1,'\n')
print(df1.set_index('C5'))
```
