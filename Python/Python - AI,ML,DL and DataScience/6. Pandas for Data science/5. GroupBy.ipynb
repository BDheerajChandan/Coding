```python
# .Groupby() - 
'''
It allows you to group together rows based off of a column and perform an aggregate function on them.
It is used for grouping the data according to the categories and applying a function to the categories. 
It also helps to aggregate data efficiently.
It is used to split the data into groups based on some criteria.
It  involves some combination of splitting the object, applying a function, and combining the results. 
The groupby is one of the most frequently used Pandas functions in data analysis. 
It is used for grouping the data points (i.e. rows) based on the distinct values in the given column or columns. 
We can then calculate aggregated values for the generated groups.
'''
```

```python
import numpy as np 
import pandas as pd
data={'Name':['Dheeraj','Chandan','Ramesh','Mohit','Suresh','Prasad'],
      'Marks':[80,78,90,94,58,77],
      'Roll':[1,2,3,4,5,6]
      }
df1=pd.DataFrame(data)
print(df1,'\n')
print(df1.groupby("Name").mean(),'\n')
print(df1.groupby("Name").max(),'\n')
info1=df1.groupby("Marks")
print(info1.prod(),'\n')
print(info1.sum(),'\n')
```

```python
import pandas as pd
data2={'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
df2=pd.DataFrame(data2)
print(df2,'\n')
print(df2.groupby("Company").mean(),'\n')
print(df2.groupby("Company").max(),'\n')
info1=df2.groupby("Company")
print(info1.prod(),'\n')
print(info1.sum(),'\n')
print(info1.std(),'\n')
print(info1.sum().loc['FB'],'\n')
print(df2.groupby("Company").sum().loc['FB'],'\n')
print(df2.groupby("Company").count(),'\n')
print(df2.groupby("Company").max(),'\n')
print(df2.groupby("Company").min(),'\n')
# .describe() - It returns description of the data in the DataFrame.
#  It is used for calculating some statistical data like percentile, mean and std of the numerical values of the Series or DataFrame. 
#  It analyzes both numeric and object series and also the DataFrame column sets of mixed data types.
print(df2.groupby("Company").describe(),'\n')
print(df2.groupby("Company").describe().T,'\n')
print(df2.groupby("Company").describe().transpose(),'\n')
print(df2.groupby("Company").describe().transpose()['FB'],'\n')
```
