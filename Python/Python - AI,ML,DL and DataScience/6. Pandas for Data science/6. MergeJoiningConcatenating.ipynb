```python
''''
There are 3 main ways of combining DataFrames together : Merging , Joining and Concatenating.
The merge() method updates the content of two DataFrame by merging them together, using the specified method(s).updates the content of two DataFrame by merging them together, using the specified method(s).
The join() method inserts column(s) from another DataFrame, or Series. 
The concat() function does all the heavy lifting of performing concatenation operations along with an axis of Pandas objects while performing optional set logic (union or intersection) of the indexes (if any) on the other axes.Concatenate pandas objects along a particular axis.

Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes whereas the merge method is more versatile and allows us to specify columns beside the index to join on for both dataframes
'''
```

```python
import pandas as pd
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                  index=[0,1,2,3])

df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                  index=[4,5,6,7])

df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                  'B':['B8','B9','B10','B11'],
                  'C':['C8','C9','C10','C11'],
                  'D':['D8','D9','D10','D11']},
                  index=[8,9,10,11])
print(df1,'\n\n',df2,'\n\n',df3)
```

```python
'''
Conatenation - It generally glues together dataframes by keeping in mind that dimensions should match along the axis you are concatenating on. pd.concat()
'''
print(pd.concat([df1,df2,df3]),'\n')
print(pd.concat([df1,df2,df3],axis=1),'\n')
print(pd.concat([df1,df2,df3],axis=0),'\n')
print(pd.concat([df1,df2,df3]).fillna(value='-'),'\n')
print(pd.concat([df1,df2,df3],axis=1).fillna(value='*'),'\n')
print(pd.concat([df1,df2,df3],axis=0).fillna(value='.'),'\n')
```

```python
left=pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'Key':['K0','K1','K2','K3']})

right=pd.DataFrame({'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3'],
                   'Key':['K0','K1','K2','K3']})

print(left,'\n\n',right,'\n\n')
'''
Merging - The merging function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.
'''
print(pd.merge(left,right),'\n')
print(pd.merge(left,right,how='inner',on='Key'),'\n')
print(pd.merge(left,right,how='right',on='Key'),'\n')
print(pd.merge(left,right,how='left',on='Key'),'\n')
```

```python
left=pd.DataFrame({'Key1':['K0','K1','K2','K3'],
                   'Key2':['K0','K1','K2','K3'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

right=pd.DataFrame({'Key1':['K0','K1','K2','K3'],
                   'Key2':['K0','K1','K2','K3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})

print(left,'\n\n',right,'\n\n')
print(pd.merge(left,right,on=['Key1','Key2']),'\n')
print(pd.merge(left,right,how='inner',on=['Key1','Key2']),'\n')
print(pd.merge(left,right,how='outer',on=['Key1','Key2']),'\n')
print(pd.merge(left,right,how='right',on=['Key1','Key2']),'\n')
```

```python
'''
.join() - Joining is a convinient method for combining the columns of two potentially differently indexed DataFrames into a single result DataFrame.
'''
left=pd.DataFrame({'A':['A0','A1','A2'],
                   'B':['B0','B1','B2']},
                   index=['k0','k1','k2'])

right=pd.DataFrame({'C':['C0','C1','C2'],
                    'D':['D0','D1','D2']},
                    index=['k0','k1','k2'])

print(left,'\n\n',right,'\n\n')
print(left.join(right),'\n')
print(left.join(right,how='outer'),'\n')
```
