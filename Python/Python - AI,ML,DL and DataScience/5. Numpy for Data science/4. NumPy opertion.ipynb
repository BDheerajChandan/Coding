```python
'''                #Trigonometric functions'''
import numpy as np

x = np.array([-1, +1, +1, -1])
y = np.array([-1, -1, +1, +1])
print("arctan2  : ",np.arctan2(x,y)) # computes element-wise arc tangent of arr1/arr2 choosing the quadrant correctly.
print("hypot : ",np.hypot(3*np.ones((3, 3)), 4*np.ones((3, 3)))) # To calculate hypotenuse for the right angled triangle
                                                                 # given its side and perpendicular.

s=[i for i in np.arange(0,361,15)]
print(s)
print("Sin of ",s," : ",np.sin(s))              # To find sine of angles
print("arcsinh of ",s,' : ',np.arcsinh(s))      # To calculate inverse hyperbolic sine
print("Cos of ",s," : ",np.cos(s))              # To calculate cosine of angles
print("Tan of ",s," : ",np.tan(s))              # To calculate Tangent of angles
print("arcsin of ",s," : ",np.arcsin(s))        # To calculate inverse sine
print("arccos of ",s," : ",np.arccos(s))        # To calculate inverse cosine
print("arctan of ",s," : ",np.arctan(s))        # To calculate inverse tangent
print("sinh of ",s," : ",np.sinh(s))            # To calculate hyperbolic sine
print("cosh of ",s," : ",np.cosh(s))            # To calculate hyperbolic cosine
print("tanh of ",s," : ",np.tanh(s))            # To calculate hyperbolic tangent
print("arccosh of ",s," : ",np.arccosh(s))      # To calculate inverse hyperbolic cosine
print("arctanh of ",s," : ",np.arctanh(s))      # To calculate inverse hyperbolic tangent
print("degree of ",s," : ",np.degrees(s))       # To find Degrees of angles
print("radians of ",s," : ",np.radians(s))      # To find Radians of angles
print("deg2rad of ",s," : ",np.deg2rad(s))      # To convert degree to radian
print("rad2deg of ",s," : ",np.rad2deg(s))      # To convert radian to degree
```

```python
''' #Bit-twiddling functions '''
import numpy as np
s=[i for i in np.arange(1,101,5)]

print(np.bitwise_and(s[::],s[::-1]))# To Compute the bit-wise AND of two array element-wise(binary representation of integers in input arrays).
print(np.bitwise_and(10,11))

print(np.bitwise_or(s[::],s[::-1]))# To Compute the bit-wise OR of two array element-wise(binary representation of integers in input arrays).
print(np.bitwise_or(10,11))

print(np.bitwise_xor(s[::],s[::-1]))# To Compute the bit-wise XOR of two array element-wise(binary representation of integers in input arrays).
print(np.bitwise_xor(11,10))

x = np.invert(np.array(13, dtype=np.uint8))# To Compute the bit-wise Inversion of an array element-wise.
                                           # It computes the bit-wise NOT of the binary representation of the integers in the input arrays.
                                           # np.uint8 -> It indicates unsigned 8 bit integer
print(x)
print(np.left_shift(5, 2))# To Shift the bits of an integer to the left.
print(np.left_shift(s, 2))

print(np.right_shift(s, 2))# To Shift the bits of an integer to the right.
print(np.right_shift(5, 2))

print(np.binary_repr(5))   # To represent binary form of the input number as a string.
print(np.binary_repr(20))
```

```python
'''                #Comparison functions
'''
import numpy as np
a=[i for i in np.arange(1,11)]
print(np.greater(12,40)) # It checks which is 'greater' out of given inputs
print(np.greater(a,5))

print(np.greater_equal(12,40)) # It checks which is 'greater or equal' to out of given inputs
print(np.greater_equal(a,5))

print(np.less(12,40)) # It checks which is 'lesser' out of given inputs
print(np.less(a,5))

print(np.less_equal(12,40)) # It checks which is 'lesser or equal' to out of given inputs
print(np.less_equal(a,5))

print(np.not_equal(12,40)) # It checks whether given inputs are 'not equal'
print(np.not_equal(a,5))

print(np.equal(12,40))# It checks whether given inputs are 'equal'
print(np.equal(a,5))

print(np.logical_and(12,40)) # It compute the AND operation on inputs.
print(np.logical_and(a,5))

print(np.logical_or(12,40)) # It compute the OR operation on inputs.
print(np.logical_or(a,5))

print(np.logical_xor(12,40)) # It compute the XOR operation on inputs.
print(np.logical_xor(a,5))

print(np.logical_not(a)) # It compute the NOT operation on input.
print(np.logical_not(5))

print(np.maximum(12,40)) # It finds the element-wise maximum of array elements.
                         # It compares two arrays and returns a new array containing the element-wise maxima.
print(np.maximum(a,5))

print(np.minimum(12,40))# It finds the element-wise minimum of array elements.
                         # It compares two arrays and returns a new array containing the element-wise minima.
print(np.minimum(a,5))

print(np.fmax(12,40))   # To compute element-wise maximum of array elements. 
                        # This function compare two arrays and returns a new array containing the element-wise maxima.
print(np.fmax(a,5))

print(np.fmin(12,40))   # To compute element-wise minimum of array elements. 
                        # This function compare two arrays and returns a new array containing the element-wise minimum.
print(np.fmin(a,5))
```

```python
'''
                     #Floating functions
'''
import numpy as np
a=[i for i in np.arange(1,11)]
print(np.isfinite(a)) #returns Boolean values True or False depending on the following: It returns False if x is positive infinity, 
                      # negative infinity or NaN. It returns True in all other cases.
print(np.nan,np.isfinite(np.nan)) 
print(np.inf,np.isfinite(np.inf))
print(np.NINF,np.isfinite(np.NINF))

print(np.isinf(a)) # checks whether a number is infinite or not. (+/- number)
x = np.array([-np.inf, 0., np.inf])
y = np.array([2, 2, 2])
print("Checking for infinity : ", np.isinf(x, y)) 

print(np.isnan(a)) # element-wise whether it is NaN or not and returns the result as a boolean array.
print(np.isnan(x,y))
print(np.isnan(1))
print(np.isnan(0))
print(np.isnan(np.nan), "\n")# not a number
print(np.isnan(np.inf), "\n")# infinity
print(np.isnan(np.NINF), "\n")  # Not a Number is not equivalent to infinity. Defines an alias for negative infinity: 






print(np.isnat(np.datetime64("Nat"))) # we can get the boolean value as true if date defined in a np.datetime64() method is not a time by 
                                        # using numpy.isnat() method.
print(np.isnat(np.datetime64("1998-01-01")))
print(np.fabs(a),',',np.fabs(10)) # used to compute the absolute values element-wise. This function returns the absolute values 
                                    #(positive magnitude) of the data in arr. It always return absolute values in floats.
print(np.signbit(a)) # element – wise check whether the signbit is set or not
print(np.copysign(1.3, -1)) # to change the sign of 2 different arrays or elements.
print(1/np.copysign(0, 1))
print(1/np.copysign(0, -1))
print(np.copysign([-1, 0, 1], -1.1))
print(np.copysign([-1, 0, 1], np.arange(3)-1))
print(np.signbit([1, -23, +34, 11]))
print(np.signbit([-1, 2, -3, -4]))
print(np.signbit(np.copysign([1,-23,+34,11],[-1,2,-3,-4])))


eps = np.finfo(np.float64).eps #finfo- shows machine limits for floating point types
print(eps)
print(np.nextafter(1, 2) == eps + 1) # to return next floating-point value after arr1 towards arr2, element-wise.
print(np.nextafter([1, 2], [2, 1]) == [eps + 1, 2 - eps])

print(np.spacing(a))
print(np.modf(a),np.modf(3.6),np.modf(-10),np.modf(10)) # returns the fractional and integer parts of the number in a two-item tuple.
print(np.ldexp(5, np.arange(4))) #This is also called as inverse of numpy.frexp() function. 
print(np.ldexp(*np.frexp(a)))

print(np.frexp(a)) # returns the mantissa and the exponent of a specified number, as a pair (m,e). 
                    # The mathematical formula for this method is: number = m * 2**e.
print(np.fmod([-3, -2, -1, 1, 2, 3], 2)) # to calculate the Module of the specified given arguments.
print(np.fmod([5, 3], [2, 2.]))

print(np.remainder([-3, -2, -1, 1, 2, 3], 2)) #It returns element-wise remainder of division.
print(np.remainder([1,2,3,4,5],[3,5,1,2,6]))

print(np.floor(a))  # returns the floor of the elements of array
print(np.ceil(a)) #  return the ceil values for each element
print(np.trunc(a)) # returns the truncated value of the elements of array.
```

```python
'''#Statistical functions
'''
import numpy as np
a=[i for i in np.arange(1,11)]
print(np.amin(a))# Return the minimum of an array or minimum along an axis
print(np.amax(a))# Return the maximum of an array or minimum along an axis
print(np.ptp(a))# Returns the range between minimum and maximum values along a specified axis. 
print(np.percentile(a, 5, axis=0))  # Used to compute the nth percentile of the given data (array elements) along the specified axis.
print(np.median(a))                     # To find median 
print(np.mean(a))                       # To find mean 
print(np.std(a))                        # To find standard deviation
print(np.var(a))                        # To find variance
print(np.average(a))                    # To find average 
```

```python
'''#Math operations:
'''
 
import numpy as np
a=[i for i in np.arange(1,11)]
print(np.add(a,9))              # To add each element with a number
print(np.subtract(2,6))# # To subtract each element with a number
print(np.multiply(a,10)) # To multiply each element with a number
print(np.matmul([[1,0],[0,1]],[[1,2],[3,4]])) # It returns the matrix product of two arrays
print(np.divide(2,5)) # It computes the element-wise division of array elements or any two numbers. 
print(np.logaddexp(a,10)) # to calculate Logarithm of the sum of exponentiations of the inputs.
print(np.logaddexp2(a,10)) # to calculate Logarithm of the sum of exponentiations of the inputs in base-2.
print(np.true_divide(a,10)) # Returns a true division of the inputs, element-wise.
                            # Array element from first array is divided by the elements from second array. 
print(np.floor_divide(a,10)) # Return the largest integer smaller or equal to the division of the inputs. 
                             # It is equivalent to the Python // operator.
print(np.negative(a))#  It is used when we want to compute the negative of array elements. It returns element-wise negative value of an array.
print(np.positive(a))# computes the positive of a number or array element-wise.It returns the element-wise numerical positive for the input array. 
print(np.power(a,3)) # Array element from first array is raised to the power of element from second element
print(np.float_power(a,0.5)) # return a usable result for negative powers and seldom overflow for +ve powers.
print(np.remainder(a,5)) # It returns element-wise remainder of division between two array 
print(np.mod(a,5)) #       It returns element-wise remainder of division between two array arr1 and arr2 i.e. arr1 % arr2 . 
print(np.fmod(a,5)) # returns an array that contains the remainder of the division of the two given input arrays.
print(np.divmod(a,5)) # returns an array showing the element-wise quotient and remainder of the division of two arrays simultaneously.
print(np.absolute(a)) #  calculates the absolute value of the values in a Numpy array. 
print(np.fabs(a)) # returns the absolute values (positive magnitude) of the data.
print(np.rint(a)) # rounds elements of the array to the nearest integer.
print(np.sign(a)) # to indicate the sign of a number element-wise
print(np.heaviside(a,5)) # used to compute the Heaviside step function of an input array passed to it.
print(np.conj(a)) # to conjugate any complex number.The conjugate of a complex number is obtained by changing the 
                                  # sign of its imaginary part. 
print(np.conjugate(a)) # o conjugate any complex number
print(np.exp(a)) #to calculate the exponential values 
print(np.exp2(a))# to calculate 2**x for all x being the array elements.
print(np.log(a)) # computes the natural logarithm of an input array's elements.
print(np.log2(a))# to calculate Base-2 logarithm of x where x belongs to all the input array elements.
print(np.log10(a)) # returns the base-10 logarithm of a number.
print(np.expm1(a)) # to calculate exponential of all the elements subtracting 1 from all the input array elements.
print(np.log1p(a)) #  a complex analytical function that has a branch cut [-inf, -1] and is continuous from above on it. 
print(np.sqrt(a))         # To find Squareroot
print(np.square(a))       # To find square
print(np.cbrt(a))         # To find cuberoot
print(np.reciprocal(a))   # To find reciprocal
print(np.gcd(a,a))        # To find HCF
print(np.lcm(a,10))       # To find LCM 
```
