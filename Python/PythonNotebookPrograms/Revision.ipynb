# Matplotlib (Scatter, pie, bar, line)

```python
# Line plot
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('mango.csv')
weight=data["WEIGHT_g"]
volume=data["VOLUME_ml"]
plt.plot(weight,marker='o',ms=10,mfc='r',mec='b')
plt.plot(volume,marker='x',ms=5,mfc='y',mec='c')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("XY-Graph")
plt.legend(['weight','volume'])
plt.show()
```

```python
# bar
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('mango.csv')
weight1=data["WEIGHT_g"][:5]
weight2=data["WEIGHT_g"][5:10]
volume=data["VOLUME_ml"]
plt.bar(range(len(weight2)),weight2)
b=plt.bar(range(len(weight1)),weight1,alpha=0.5)
c=['r','b','g','c','y']
cnt=0
for i in c:
    b[cnt].set_color(i)
    cnt=cnt+1
plt.legend(b,c)
plt.show()
```

```python
# Scatter plot
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('mango.csv')
data
weight=data["WEIGHT_g"][:5]
volume=data["VOLUME_ml"][:5]
plt.scatter(range(len(weight)),weight,marker='o',s=100,c='y',edgecolors='r',facecolors='k')
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('mango.csv')
Clabel=['r','b','g','c','y']
weight=data["WEIGHT_g"][:5]
volume=data["VOLUME_ml"][:5]
plt.pie(weight,startangle=0,labels=Clabel)
plt.legend(Clabel)
plt.show()
```

```python
# Factorial recurrsion
def  fn(n): 
    if(n==1):
        return 1
    else: 
        return n*fn(n-1)
fn(5)
```

```python
def fibo(n):
    if n<=1:
        return n
    else:
        return fibo(n-1)+fibo(n-2)
fibo(5)
```
