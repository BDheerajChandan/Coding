# Line plot - plt.plot()

## single plot - 1 graph

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("mango.csv")
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
plt.plot(weight)
plt.plot(shortW)
plt.plot(longW)
plt.plot(distance)
plt.legend(plotNames)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Graph")
plt.xlim(0,250)
plt.ylim(0,500)
plt.plot(weight,marker='o',ms=10,mec='r',mfc='c',linestyle='dotted')
plt.show()
```

# Subplot - plt.subplot()

## multiple plot - 4 graph

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("mango.csv")
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
plt.figure(figsize=[10,10])
plt.suptitle("Mango properties")
plt.subplot(2, 2, 1)
plt.plot(weight,c='r')
plt.xlabel("X Axis")
plt.ylabel("Y axis")
plt.title("weight")
plt.subplot(2, 2, 2)
plt.plot(shortW,c='g')
plt.xlabel("X Axis\n")
plt.ylabel("Y axis")
plt.title("shortW")
plt.subplot(2, 2, 3)
plt.plot(longW,c='b')
plt.xlabel("X Axis")
plt.ylabel("Y axis")
plt.title("longW")
plt.subplot(2, 2, 4)
plt.plot(distance,c='c')
plt.xlabel("X Axis")
plt.ylabel("Y axis")
plt.title("distance")
plt.show()
```

# bar plot - plt.bar()

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("mango.csv")
weight=data["WEIGHT_g"]
shortW=data["DIA1_mm_shortwidth"]
longW=data["DIA2_mm_longwidth"]
distance=data["Distance"]
plotNames=["weight","shortW","longW","distance"]
clr=['r','b','g','y','c']
b=plt.bar(range(len(weight[:5])),weight[:5],alpha=0.5)
cnt=0
for i in clr:
    b[cnt].set_color(i)
    cnt=cnt+1
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("XY Graph")
plt.ylim(0,max(weight[:5])+50)
plt.legend(b,clr)
plt.grid(axis='x')
plt.grid(axis='y')
plt.plot(weight[:5],marker='o',ms=10,mec='r',mfc='c',linestyle='dotted')
plt.yticks(range(0,301,25))
plt.xticks(range(0,6,1))
plt.show()
```
